**Q3.** Display the total number of books written by each author, ordered by count (descending).

```python
# Join the two table (authors and books) 
authors_books = pd.merge(df_authors, df_books, on="author_id")
authors_books
```

```python
authors_books.columns
```

```python
# GROUP BY book id  # COUNT by book_id,and then Rename column and then order it by descending
Tot_no_bk_written = authors_books.groupby(
    'author_name')['book_id'].count().reset_index(name='total_books').sort_values('total_books', ascending=False)

Tot_no_bk_written
```

**Q12.** Generate a report showing monthly borrowing trends for the past year (count of books borrowed per month).

```python
df_borrowhistory.head(1)
```

```python
df_borrowhistory.columns
```

```python
# Convert borrow_date to datetime
import pandas as pd
import datetime

df_borrowhistory['borrow_date'] = pd.to_datetime(df_borrowhistory['borrow_date'])
df_borrowhistory.head(1)
```

```python
print(df_borrowhistory['borrow_date'].min())
print(df_borrowhistory['borrow_date'].max())
```

```python

# Convert borrow_date to datetime
df_borrowhistory['borrow_date'] = pd.to_datetime(df_borrowhistory['borrow_date'])

# Using borrowhistory max date filter
max_date = df_borrowhistory['borrow_date'].max()
one_year_ago = max_date - pd.DateOffset(years=1)

# Filter last 12 months relative to max date
filtered_df = df_borrowhistory[df_borrowhistory['borrow_date'] >= one_year_ago]

# Group by month
Monthly_borrowed_trend = (
    filtered_df
    .groupby(filtered_df['borrow_date'].dt.to_period('M'))
    .size()
    .reset_index(name='total_borrowed')
)

print(Monthly_borrowed_trend)
```

**Q14.** Calculate the total revenue from book orders per supplier, showing only suppliers with orders exceeding $5,000.

```python
df_bookorder.columns
```

```python
df_bookorder['o_cost'].unique()
```

```python
# Calculate revenue per row
df_bookorder['Revenue'] = df_bookorder['quantity'] * df_bookorder['o_cost']
df_bookorder['Revenue']
```

```python
# Group Revenue by supplier and filter
Tot_rev_per_supplier = (
    df_bookorder.groupby('supplier_name')['Revenue']
    .sum()
    .reset_index(name='total_revenue')
    .sort_values('total_revenue', ascending=False)
)

print(Tot_rev_per_supplier)
```

```python
supplier_lessthan_60_rev = Tot_rev_per_supplier.query('total_revenue < 60') 
supplier_lessthan_60_rev
```

```python
supplier_morethan_5000_rev = Tot_rev_per_supplier.query('total_revenue > 5000') 
supplier_morethan_5000_rev
```
